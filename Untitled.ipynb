```python
params = []
comps = nx.connected_components(g)
for component in comps:
    nodedegree = []
    nodeeccentricity = []
    nodediameter = []
    noderadius = []
    sub = g.subgraph(component)
    params.append(nx.number_of_nodes(sub))
    params.append(nx.number_of_edges(sub))
    sub_nodes = nx.nodes(sub)
    for u in sub_nodes:
        nodedegree.append(sub.degree(u))
        nodeeccentricity.append(nx.eccentricity(sub, v=u))
    params.append(nodedegree)
    params.append(nx.eccentricity(sub))
    params.append(min(nodeeccentricity)) 
    params.append(max(nodeeccentricity)) 
params = np.array(params, dtype=object)
params = params.reshape(5,6)
```
